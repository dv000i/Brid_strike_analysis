```python
from datetime import datetime
def extract_year(date_string):
    if '/' in date_string:
        # If date string uses '/', use a different format
        date_object = datetime.strptime(date_string, '%m/%d/%y %H:%M')
    else:
        # If date string uses '-', use another format
        date_object = datetime.strptime(date_string, '%m-%d-%Y %H:%M')
    return date_object.year

# Extracting years from each date string
years = [extract_year(date_str) for date_str in date_strings]

```
